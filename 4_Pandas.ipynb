```python
datos=[["Helena", "F","Bucaramanga","13"],
       ["Esteban","M","Medellin","27"],
       ["Edith","F","Bucaramanga","59"],
       ["Jairo","M","Cúcuta","58"],
       ["José","M","Cali","45"]]
```

```python
datos
```

```python
titulos=["Nombre","Genero","Ciudad","Edad"]
indices=["uno","dos","tres","cuatro","cinco"]
```

```python
df=pd.DataFrame(datos,index=indices, columns=titulos)
df
```

```python
df["Telefono"]=["3213213211","3131313133","3123123122","3143143144","3003003001"]
df
```

```python
df["Edad"].dtype
```

```python
df["Edad"]=pd.to_numeric(df["Edad"])
```

```python
df["Edad"].dtype
```

```python
df[(df["Edad"]<30)]  #Filtramos para ver la tabla con los datos de personas menores de 30
```

```python
df[(df["Ciudad"]=="Bucaramanga")] #Filtramos para ver personas que esten en Bucaramanga
```

```python
df[(df["Edad"]<30) & (df["Ciudad"]=="Bucaramanga")] #Filtramos doble, personas que tienen menos de 30 años y estan en bBUcaramanga
```

```python
ciudad=input("Ingrese lo que desea buscar: ")  #Busquemos valores en una columna
df[df.Ciudad.str.contains(ciudad)]
```

```python
df["Ciudad"]=df["Ciudad"].replace("Medellin","Medellín")  #Corrigiendo un valor en una columna de la tabla
df
```
