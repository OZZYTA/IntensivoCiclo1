Se solicita una funcion que calcule 3 tipos de operaciones: 
*   O
*   Y
*   MULT
.

Dicha calculadora solo recibira dos cadenas de texto (listas) de la misma longitud, compuestas por caracteres A Y B, y la operación que se requiere realizar. Y su resultado sera, segun corresponda a la operación, como esta demostrado en la siguiente tabla, una A o una B. 

calculadora.jpg

La salida o resultado debe ser impreso igualmente como una cadena de texto que tendra la misma medida de las cadenas originales.

EJEMPLO:

lista1 : 'AAB'

Lista2: 'ABA'

Operacion: 'Y'

Resultado: 'AAA'

lista1=BBA
lista2=ABA
Operacion: MULT
RESULTADO:BAA

````python
def calculador(lista1,lista2,operacion):
  resultado=""
  if len(lista1)==len(lista2):
    if operacion=="O":
      for i in range(len(lista1)):
        if lista1[i]=="A" and lista2[i]=="A":
          resultado=resultado+"A"
        elif lista1[i]=="A" and lista2[i]=="B":
          resultado=resultado+"B"
        elif lista1[i]=="B" and lista2[i]=="A":
          resultado=resultado+"A"
        elif lista1[i]=="B" and lista2[i]=="B":
          resultado=resultado+"B"
    elif operacion=="Y":
      for i in range(len(lista1)):
        if lista1[i]=="A" and lista2[i]=="A":
          resultado=resultado+"A"
        elif lista1[i]=="A" and lista2[i]=="B":
          resultado=resultado+"A"
        elif lista1[i]=="B" and lista2[i]=="A":
          resultado=resultado+"A"
        elif lista1[i]=="B" and lista2[i]=="B":
          resultado=resultado+"B"
    elif operacion=="MULT":
      for i in range(len(lista1)):
        if lista1[i]=="A" and lista2[i]=="A":
          resultado=resultado+"A"
        elif lista1[i]=="A" and lista2[i]=="B":
          resultado=resultado+"B"
        elif lista1[i]=="B" and lista2[i]=="A":
          resultado=resultado+"B"
        elif lista1[i]=="B" and lista2[i]=="B":
          resultado=resultado+"A"
    return resultado
  else:
    return "Dimensiones de las listas no coinciden"
````
